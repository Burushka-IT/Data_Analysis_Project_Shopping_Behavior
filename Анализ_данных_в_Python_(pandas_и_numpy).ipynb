Выбросы

```python


    # Выбросы (Z-Score с использованием NumPy)
print("Анализ выбросов (Z-Score через NumPy):")
numerical_col = 'Purchase Amount (USD)'

    # Расчет среднего и стандартного отклонения с помощью numpy
mean_val = np.mean(df[numerical_col])
std_val = np.std(df[numerical_col])

    # Расчет Z-Score вручную
df['Z_Score_Purchase'] = (df[numerical_col] - mean_val) / std_val

    # Определение выбросов (Z-Score > 3 или < -3)
outliers = df[(df['Z_Score_Purchase'] > 3) | (df['Z_Score_Purchase'] < -3)]
print(f"Количество выбросов в '{numerical_col}' (Z-Score > 3): {len(outliers)}")

df.drop(columns=['Z_Score_Purchase'], inplace=True) # Удаляем вспомогательный столбец
print("Анализ выбросов завершен.")
```

```python
print("\n5. Корреляционный анализ:")
numeric_cols = ['Age', 'Purchase Amount (USD)', 'Review Rating', 'Previous Purchases']
correlation_matrix = df[numeric_cols].corr()
print("Матрица корреляции:")
print(correlation_matrix)
```

Визуализация данных (5 графиков)

```python
# Числовые столбцы для корреляции для heatmap
numeric_cols = ['Age', 'Purchase Amount (USD)', 'Review Rating', 'Previous Purchases']
correlation_matrix = df[numeric_cols].corr()

plt.figure(figsize=(18, 16))

    # 1. Гистограмма распределения
plt.subplot(3, 2, 1)
sns.histplot(df['Purchase Amount (USD)'], kde=True, bins=30, color='skyblue')
plt.title('1. Распределение Суммы Покупки (Purchase Amount)')
plt.xlabel('Сумма Покупки (USD)')

    # 2. Boxplot
plt.subplot(3, 2, 2)
sns.boxplot(y=df['Purchase Amount (USD)'], color='lightcoral')
plt.title('2. Boxplot Суммы Покупки (для проверки выбросов)')
plt.ylabel('Сумма Покупки (USD)')

    # 3. Столбчатая диаграмма (Покупки по категории)
plt.subplot(3, 2, 3)
category_counts = df['Category'].value_counts()
sns.barplot(x=category_counts.index, y=category_counts.values, palette='viridis')
plt.title('3. Количество Покупок по Категории')
plt.xlabel('Категория Товара')
plt.ylabel('Количество Покупок')
plt.xticks(rotation=45, ha='right') # Поворот подписей для читаемости

    # 4. Линейный график (Средняя сумма покупки по возрасту)
plt.subplot(3, 2, 4)
age_purchase = df.groupby('Age')['Purchase Amount (USD)'].mean()
age_purchase.plot(kind='line', marker='o', color='forestgreen', linewidth=2)
plt.title('4. Средняя Сумма Покупки в Зависимости от Возраста')
plt.xlabel('Возраст')
plt.ylabel('Средняя Сумма Покупки (USD)')

    # 5. Heatmap корреляций
plt.subplot(3, 2, (5, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('5. Тепловая Карта Корреляций Числовых Переменных')

plt.tight_layout(pad=3.0)
plt.show()
```
